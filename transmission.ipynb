```python
features = [
    "Gear_Slip_Frequency",
    "Clutch_Wear_Percentage",
    "Transmission_Temp",
    "Axle_Wear_Score"
]

for col in features:
    plt.boxplot(df_transmission[col].dropna())
    plt.title(col)
    plt.show()
```

```python
feature_cols = [
    "Clutch_Wear_Percentage",
    "Transmission_Temp",
    "Axle_Wear_Score"
    
]
df_transmission['Gear_Slip_Frequency']=df_transmission['Gear_Slip_Frequency'].fillna(df_transmission['Gear_Slip_Frequency'].mode()[0])
for col in feature_cols:
    df_transmission[col] = df_transmission[col].fillna(df_transmission[col].median())
```

```python
df_transmission.isna().sum()
```

```python
x = df_transmission.drop(
    columns=["Transmission_Remaining_Years", "Car_ID"]
)
y = df_transmission["Transmission_Remaining_Years"]
```

```python
for col in x.columns:
    lower = x[col].quantile(0.01)
    upper = x[col].quantile(0.99)
    x[col] = x[col].clip(lower, upper)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.3,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(
    n_estimators=250,
    max_depth=9,
    min_samples_leaf=12,
    random_state=42
)

model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("MAE :", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2  :", r2_score(y_test, y_pred))
```

```python
print("Train R2:", model.score(x_train, y_train))
print("Test R2 :", model.score(x_test, y_test))
```

```python
import joblib

joblib.dump(model, "transmission_model.pkl")
```

```python
df_transmission["Predicted_Transmission_Remaining_Years"] = model.predict(
    df_transmission.drop(columns=["Car_ID", "Transmission_Remaining_Years"])
)
```

```python
def Transmission_maintenance_status(years):
    if years < 2:
        return "Immediate Repair"
    elif years < 4:
        return "Service Soon"
    else:
        return "Healthy"
```

```python
df_transmission.loc[:, "Transmission_Maintenance_Status"] = (
    df_transmission["Predicted_Transmission_Remaining_Years"]
    .apply(Transmission_maintenance_status)
)
```
